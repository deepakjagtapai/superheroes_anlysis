a.	Find names, real names and alias of superhero who is part of most teams.

````python
df["name"].fillna("NA", inplace=True)
````

````python
creator_counts = df['creator'].value_counts()
count_df = pd.DataFrame({'count': creator_counts})
count_df.reset_index(inplace=True)
count_df.rename(columns={"index":"creator"},inplace=True)
````

````python
df['aliases']=df['aliases'].apply(ast.literal_eval)
````

````python
df['teams']=df['teams'].apply(ast.literal_eval)
````

````python
other_cols = df.groupby('creator').agg({'name': list, 'real_name': list,'aliases':list,'teams':list}).reset_index()
final = count_df.merge(other_cols, on="creator")
final.head()
````

````python
def flatten_and_remove_empty(lst):
    flattened = []
    for item in lst:
        if isinstance(item, list):
            flattened.extend(flatten_and_remove_empty(item))
        elif item.strip():
            flattened.append(item)
    return flattened

final['aliases'] = final['aliases'].apply(flatten_and_remove_empty)
final['teams'] = final['teams'].apply(flatten_and_remove_empty)
final.head(10)
````

b.	Are there any crossovers between creators and teams?

````python
def converttolist(text):
  text=list(text)
  return text
final["name"]=final["name"].apply(converttolist)
````

````python
def flatten(test_list):
    if isinstance(test_list, list):
        temp = []
        for ele in test_list:
            temp.extend(flatten(ele))
        return temp
    else:
        return [test_list]
final['name'] = final['name'].apply(flatten)
````

````python
unique_names = set()

for index, row in final.iterrows():
    creator = row['creator']
    names = row['teams']
    name_set = set(names)
    common_elements = name_set.intersection(unique_names)

    if common_elements:
        print(f"Creator '{creator}' has crossovers with Teams: {common_elements}")

    unique_names.update(name_set)
````

````python
import torch
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer
````

````python
model_path = 'cardiffnlp/twitter-xlm-roberta-base-sentiment'
model = AutoModelForSequenceClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)
````

````python
def analyze_sentiment(text):
    tokens = tokenizer(text, truncation=True, padding='longest', return_tensors='pt', max_length=512)
    with torch.no_grad():
        outputs = model(**tokens)
    label = torch.argmax(outputs.logits, dim=1).item()
    sentiment_label = 'positive' if label == 1 else 'negative'

    return sentiment_label
````

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
model.to(device)
````

````python
df_tensor = df.applymap(lambda x: x.to(device) if torch.is_tensor(x) else x)
df_tensor['past_sentiment'] = df_tensor['history_text'].apply(analyze_sentiment)
````
